```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values (only for x > 0)
x = np.linspace(0, 50, 1000)

# Define piecewise function
y = np.piecewise(x, [x < 20, x >= 20], [lambda x: x / 20, 1])

# Create the plot
plt.plot(x, y, label=r"$y=\frac{x}{20}$ if $x<20$, $y=1$ if $x\geq20$", color="blue")

# Customize the plot
plt.axhline(0, color='black', linewidth=1)  # x-axis
plt.axvline(0, color='black', linewidth=1)  # y-axis
plt.grid(True, linestyle='--', alpha=0.6)   # Grid with dashed lines
plt.xlabel("x")
plt.ylabel("y")
plt.title("Piecewise Function Plot")
plt.legend()

# Show the plot
plt.show()
```

```python
#y = 1-2^-x
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 50, 1000)
y = 1 - 2 ** -x
plt.plot(x, y, label=r"$y= 1-2^-x$", color ="blue")

plt.ylabel('y')
plt.xlabel('x')
plt.title("Exponential Function")
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.show()
```

```python
#y = tanhx
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 50, 1000)
y = np.tanh(x)
plt.plot(x, y, label=r"$y= tanh(x)$", color ="blue")

plt.ylabel('y')
plt.xlabel('x')
plt.title("tanh Function")
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.show()
```
